```python
def read_large_file(file_object) : 
    """ a generator function to lazily read a large file """

    while True : 
        data = file_object.readline()
        
        if not data : 
            break
        
        yield data
```

```python
with open(file_path) as data : 
    gen = read_large_file(data)
    print(next(gen))
    print(next(gen))
    print(next(gen))
    print(next(gen))
    print(next(gen))
    print(next(gen))
    print(next(gen))
    print(next(gen))
```
